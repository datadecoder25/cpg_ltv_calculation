## Calculating ProductRaw

```python
# Ensure purchase_date is datetime objects for proper filtering and min operations
# Use utc=True to handle mixed time zones and errors='coerce' to handle parsing errors
combined_df['purchase_date'] = pd.to_datetime(combined_df['purchase_date'], utc=True, errors='coerce')

# Remove rows where purchase_date couldn't be parsed (NaT values)
combined_df = combined_df.dropna(subset=['purchase_date'])

filtered_df = combined_df[
    (combined_df['currency'] == 'USD') &
    (combined_df['item_price'] > 0) &
    (combined_df['purchase_date'].dt.strftime('%Y-%m-01') > '2022-09-01')
].copy() # .copy() to avoid SettingWithCopyWarning

# Step 2: Calculate MIN(purchase_date) partitioned by buyer_email, merchant_sku
# Convert purchase_date to 'YYYY-MM-DD' string format as in SQL's STRFTIME
filtered_df['buyer_email'] = filtered_df['buyer_email'].fillna('NA')
filtered_df['pome_date'] = filtered_df.groupby(['buyer_email', 'merchant_sku'], dropna=False)['purchase_date'].transform('min').dt.strftime('%Y-%m-%d')

# Step 3: Select distinct buyer_email, merchant_sku, and the calculated pome_date
result_df = filtered_df[['buyer_email', 'merchant_sku', 'pome_date']].drop_duplicates().reset_index(drop=True)

# Rename buyer_email to user_id as per SQL alias
ntb = result_df.rename(columns={'buyer_email': 'user_id'})

# Select and rename columns, and perform type conversions
raw_data_df = filtered_df[[
    'buyer_email',
    'amazon_order_id',
    'merchant_sku',
    'purchase_date',
    'item_price',
    'shipped_quantity'
]].copy()

# Rename columns as per the SQL aliases
raw_data_df = raw_data_df.rename(columns={
    'buyer_email': 'user_id',
    'amazon_order_id': 'order_id',
    'item_price': 'sales',
    'shipped_quantity': 'quantity'
})

# Apply STRFTIME( "%Y-%m-%d" , purchase_date )
raw_data_df['purchase_date'] = raw_data_df['purchase_date'].dt.strftime('%Y-%m-%d')

# Cast item_price as decimal(10, 2) - Pandas handles floats with high precision
# For displaying exactly two decimal places, you might format it later if needed,
# but the underlying data type remains float.
raw_data_df['sales'] = raw_data_df['sales'].astype(float).round(2)

# Apply DISTINCT
# The SQL query uses DISTINCT on all selected columns.
raw_data = raw_data_df.drop_duplicates()
```

```python
# Convert the SQL query to pandas operations
import math

# Perform left join between raw_data and ntb
process_data = raw_data.merge(
    ntb,
    on=['user_id', 'merchant_sku'],
    how='left'
)

# Convert date columns to datetime for calculation
process_data['purchase_date'] = pd.to_datetime(process_data['purchase_date'])
process_data['pome_date'] = pd.to_datetime(process_data['pome_date'])

# Calculate date difference (equivalent to JULIANDAY difference)
process_data['date_diff'] = (process_data['purchase_date'] - process_data['pome_date']).dt.days

# Calculate months (equivalent to CEIL(date_diff / 30))
# Calculate months (equivalent to CEIL(date_diff / 30))
# Handle NaN values by filling them before converting to int
# process_data['months'] = np.ceil(process_data['date_diff'] / 30) #.astype(int)
process_data['months'] = np.ceil(process_data['date_diff'] / 30) 

# Group by user_id, merchant_sku, order_id, purchase_date, pome_date
# and aggregate sales and quantity
process_data_agg = process_data.groupby([
    'user_id', 
    'merchant_sku', 
    'order_id', 
    'purchase_date', 
    'pome_date'
], dropna=False).agg({
    'sales': 'sum',
    'quantity': 'sum',
    'date_diff': 'first',  # These should be the same within each group
    'months': 'first'      # These should be the same within each group
}).reset_index()

result = process_data_agg.groupby(['merchant_sku', 'months'], dropna=False).agg({
    'user_id': 'nunique',      # COUNT(distinct user_id) as users
    'sales': 'sum',            # SUM(sales) as sales
    'quantity': 'sum',         # SUM(quantity) as quantity
    'order_id': 'nunique'      # COUNT(distinct order_id) as orders
}).reset_index()

# Rename columns to match SQL aliases
ProductRaw = result.rename(columns={
    'user_id': 'users',
    'order_id': 'orders'  # Changed from 'users' to 'orders' - assuming this was a typo
})

print(ProductRaw.shape)
ProductRaw.head(20)
```

```python
# Step 2: Create cte_test equivalent using UNION logic
# First part: magnitude = 1 (original users/sales/quantity)
cte_test_part1 = cte_tbl[cte_tbl['magnitude'] == 1].groupby('merchant_sku', dropna=False).agg({
    'buyer_email': 'nunique',    # users
    'item_price': 'sum',         # sales  
    'shipped_quantity': 'sum'    # quantity
}).reset_index()

cte_test_part1.columns = ['merchant_sku', 'users', 'sales', 'quantity']
cte_test_part1['users1'] = 0
cte_test_part1['sales1'] = 0
cte_test_part1['quantity1'] = 0
cte_test_part1['orders'] = cte_test_part1['users']  # count distinct buyer_email as orders
cte_test_part1['orders1'] = 0

# Second part: magnitude > 1 (repeat users/sales/quantity)
cte_test_part2 = cte_tbl[cte_tbl['magnitude'] > 1].groupby('merchant_sku', dropna=False).agg({
    'buyer_email': 'nunique',    # users1
    'item_price': 'sum',         # sales1
    'shipped_quantity': 'sum'    # quantity1
}).reset_index()

cte_test_part2.columns = ['merchant_sku', 'users1', 'sales1', 'quantity1']
cte_test_part2['users'] = 0
cte_test_part2['sales'] = 0
cte_test_part2['quantity'] = 0
cte_test_part2['orders'] = 0
cte_test_part2['orders1'] = cte_test_part2['users1']  # count distinct buyer_email as orders1

# Combine both parts (equivalent to UNION)
cte_test = pd.concat([cte_test_part1, cte_test_part2], ignore_index=True)

print("\nFirst 10 rows of cte_test:")
cte_test.head(10)
```

```python
# Step 3: Final aggregation (equivalent to final SELECT)
ProductSummary = cte_test.groupby('merchant_sku', dropna=False).agg({
    'users': 'sum',
    'sales': 'sum', 
    'quantity': 'sum',
    'users1': 'sum',
    'sales1': 'sum',
    'quantity1': 'sum',
    'orders': 'sum',
    'orders1': 'sum'
}).reset_index()

ProductSummary['orders'] = ProductSummary['quantity']
ProductSummary['orders1'] = ProductSummary['quantity']

print("\nFirst 10 rows of ProductSummary:")
ProductSummary.head(20)
```

## Calculating RawData

```python
# Convert SQL RawData query to pandas operations

# Step 1: Create ntb equivalent
ntb_df = combined_df[combined_df['purchase_date'].dt.strftime('%Y-%m-01') > '2022-09-01'].copy()
ntb_df['buyer_email'] = ntb_df['buyer_email'].fillna('NA')
ntb_df['pome_month'] = ntb_df.groupby('buyer_email', dropna=False)['purchase_date'].transform(lambda x: x.dt.strftime('%Y-%m-01').min())
ntb = ntb_df[['buyer_email', 'pome_month']].drop_duplicates().rename(columns={'buyer_email': 'user_id'})

# Step 2: Create pntb equivalent  
pntb_df = combined_df[combined_df['purchase_date'].dt.strftime('%Y-%m-01') > '2022-09-01'].copy()
pntb_df['pntb_title'] = pntb_df.groupby('merchant_sku', dropna=False)['title'].transform('min')
pntb_df['pntb_month'] = pntb_df.groupby('merchant_sku', dropna=False)['purchase_date'].transform(lambda x: x.dt.strftime('%Y-%m-01').min())
pntb = pntb_df[['merchant_sku', 'pntb_title', 'pntb_month']].drop_duplicates().rename(columns={'merchant_sku': 'tracking_id'})

# Step 3: Create output_tbl equivalent
# Filter the main dataframe
filtered_df = combined_df[
    (combined_df['currency'] == 'USD') &
    (combined_df['item_price'] > 0) &
    (combined_df['purchase_date'].dt.strftime('%Y-%m-01') > '2022-09-01')
].copy()

# Create month column
filtered_df['month'] = filtered_df['purchase_date'].dt.strftime('%Y-%m-01')
filtered_df['buyer_email'] = filtered_df['buyer_email'].fillna('NA')

# Left join with ntb
output_tbl1 = filtered_df.merge(
    ntb,
    left_on='buyer_email',
    right_on='user_id',
    how='left'
)

# Left join with pntb
output_tbl = output_tbl1.merge(
    pntb,
    left_on='merchant_sku', 
    right_on='tracking_id',
    how='left'
)

output_tbl['buyer_email'] = output_tbl['buyer_email'].fillna('NA')
output_tbl['amazon_order_id'] = output_tbl['amazon_order_id'].fillna('NA')


# Group by and aggregate
RawData = output_tbl.groupby(['month', 'merchant_sku', 'pntb_title', 'pome_month', 'pntb_month'], dropna=False).agg({
    'buyer_email': 'nunique',       # users
    'shipped_quantity': 'sum',      # quantity  
    'amazon_order_id': 'nunique',   # orders
    'item_price': 'sum'             # sales
}).reset_index()

# Rename columns
RawData = RawData.rename(columns={
    'buyer_email': 'users',
    'shipped_quantity': 'quantity',
    'amazon_order_id': 'orders', 
    'item_price': 'sales'
})

# Calculate derived columns (cast as decimal(10,2) equivalent)
RawData['avg_per_item'] = (RawData['sales'] / RawData['quantity']).round(2)
RawData['avg_per_order'] = (RawData['sales'] / RawData['orders']).round(2)
RawData['avg_per_user'] = (RawData['sales'] / RawData['users']).round(2)

# Round sales to 2 decimal places (equivalent to cast as decimal(10,2))
RawData['sales'] = RawData['sales'].round(2)

RawData.head(10)
```

```python
# Create Customer Repeat Rates table structure
import pandas as pd
import numpy as np

# Define the metrics for each cohort
metrics = [
    'Active Customers',
    'Quantities', 
    'Orders',
    'Revenue',
    'Cumulative Revenue',
    'Retention Rate',
    'LTV'
]

# Get the months for columns (all months from our analysis)
month_columns = sorted(all_months)

# Create the base structure
table_data = []

# For each cohort month
for idx, row in cohort_df.iterrows():
    pome_month = row['POME Month']
    cohort_size = row['Cohort Size']
    
    # For each metric
    for metric in metrics:
        # Create a row for this cohort-metric combination
        row_data = {
            'POME Month': pome_month,
            'Cohort Size': cohort_size,
            'Metric': metric
        }
        
        # Add empty columns for each month
        for month in month_columns:
            row_data[month] = 0  # Initialize with 0, will be calculated later
        
        # Add Total column
        row_data['Total'] = 0
        
        table_data.append(row_data)

# Create the Customer Repeat Rates DataFrame
customer_repeat_rates = pd.DataFrame(table_data)

customer_repeat_rates.head(10)
```

```python
# Create a copy of the customer_repeat_rates table to work with
filled_table = customer_repeat_rates.copy()

# Convert month columns and Total column to float dtype to avoid dtype incompatibility warning
for month_col in month_columns:
    filled_table[month_col] = filled_table[month_col].astype(float)
filled_table['Total'] = filled_table['Total'].astype(float)

# Define which column to sum for each metric
metric_columns = {
    'Active Customers': 'users',
    'Quantities': 'quantity', 
    'Orders': 'orders',
    'Revenue': 'sales',
}

# Loop through each row in the table
for idx, row in filled_table.iterrows():
    pome_month = row['POME Month']
    cohort_size = row['Cohort Size']
    metric = row['Metric']
    
    # Skip if cohort size is 0 to avoid division by zero
    if cohort_size == 0:
        continue
    
    # Get the column to sum for this metric (skip Active Customers, Quantities, Orders, and Revenue as they're handled separately)
    if metric in metric_columns and metric not in ['Active Customers', 'Quantities', 'Orders', 'Revenue']:
        column_to_sum = metric_columns[metric]
        
        # Initialize running sum for cumulative metrics
        cumulative_sum = 0
        
        # Loop through each month column
        for month_col in month_columns:
            # Compare month column with POME Month
            if month_col < pome_month:
                # If month column is less than POME Month, set to 0
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise, filter RawData and sum the appropriate column
                filtered_data = RawData[
                    (RawData['month'] == month_col) & 
                    (RawData['pome_month'] == pome_month) & 
                    (RawData['IsFiltered'] == 1)
                ]
                
                # Sum the appropriate column for this condition
                value_sum = filtered_data[column_to_sum].sum()
                
                # Handle different metric calculations
                # (Orders and Revenue are now handled separately)
                pass

# Handle Active Customers calculation separately (Excel-like formula)
# Formula: =IF(E$8<$B9,"",SUMIFS(RawData!$F:$F,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1))
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Active Customers':
        pome_month = row['POME Month']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If E$8 < $B9: month column is less than POME Month, set to 0 (equivalent to "")
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise: SUMIFS(RawData!$F:$F,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1)
                # F:F = users column to sum
                # A:A = month column, E$8 = current month (month_col)
                # D:D = pome_month column, $B9 = POME Month (pome_month)
                # M:M = IsFiltered column = 1
                filtered_data = RawData[
                    (RawData['month'] == month_col) &       # A:A = month, E$8 = current month
                    (RawData['pome_month'] == pome_month) & # D:D = pome_month, $B9 = POME Month
                    (RawData['IsFiltered'] == 1)            # M:M = IsFiltered = 1
                ]
                
                # Sum the users column (F:F in Excel)
                active_customers = filtered_data['users'].sum()
                filled_table.at[idx, month_col] = round(active_customers, 0)

# Handle Quantities calculation separately (Excel-like formula)
# Formula: =IF(E$8<$B9,"",SUMIFS(RawData!$G:$G,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1))
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Quantities':
        pome_month = row['POME Month']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If E$8 < $B9: month column is less than POME Month, set to 0 (equivalent to "")
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise: SUMIFS(RawData!$G:$G,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1)
                # G:G = quantity column to sum
                # A:A = month column, E$8 = current month (month_col)
                # D:D = pome_month column, $B9 = POME Month (pome_month)
                # M:M = IsFiltered column = 1
                filtered_data = RawData[
                    (RawData['month'] == month_col) &       # A:A = month, E$8 = current month
                    (RawData['pome_month'] == pome_month) & # D:D = pome_month, $B9 = POME Month
                    (RawData['IsFiltered'] == 1)            # M:M = IsFiltered = 1
                ]
                
                # Sum the quantity column (G:G in Excel)
                quantities = filtered_data['quantity'].sum()
                filled_table.at[idx, month_col] = round(quantities, 0)

# Handle Orders calculation separately (Excel-like formula)
# Formula: =IF(E$8<$B9,"",SUMIFS(RawData!$H:$H,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1))
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Orders':
        pome_month = row['POME Month']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If E$8 < $B9: month column is less than POME Month, set to 0 (equivalent to "")
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise: SUMIFS(RawData!$H:$H,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1)
                # H:H = orders column to sum
                # A:A = month column, E$8 = current month (month_col)
                # D:D = pome_month column, $B9 = POME Month (pome_month)
                # M:M = IsFiltered column = 1
                filtered_data = RawData[
                    (RawData['month'] == month_col) &       # A:A = month, E$8 = current month
                    (RawData['pome_month'] == pome_month) & # D:D = pome_month, $B9 = POME Month
                    (RawData['IsFiltered'] == 1)            # M:M = IsFiltered = 1
                ]
                
                # Sum the orders column (H:H in Excel)
                orders = filtered_data['orders'].sum()
                filled_table.at[idx, month_col] = round(orders, 0)

# Handle Revenue calculation separately (Excel-like formula)
# Formula: =IF(E$8<$B9,"",SUMIFS(RawData!$I:$I,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1))
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Revenue':
        pome_month = row['POME Month']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If E$8 < $B9: month column is less than POME Month, set to 0 (equivalent to "")
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise: SUMIFS(RawData!$I:$I,RawData!$A:$A,E$8,RawData!$D:$D,$B9,RawData!$M:$M,1)
                # I:I = sales column to sum
                # A:A = month column, E$8 = current month (month_col)
                # D:D = pome_month column, $B9 = POME Month (pome_month)
                # M:M = IsFiltered column = 1
                filtered_data = RawData[
                    (RawData['month'] == month_col) &       # A:A = month, E$8 = current month
                    (RawData['pome_month'] == pome_month) & # D:D = pome_month, $B9 = POME Month
                    (RawData['IsFiltered'] == 1)            # M:M = IsFiltered = 1
                ]
                
                # Sum the sales column (I:I in Excel)
                revenue = filtered_data['sales'].sum()
                filled_table.at[idx, month_col] = int(round(revenue, 0))

# Handle Cumulative Revenue calculation separately (Excel-like formula)
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Cumulative Revenue':
        pome_month = row['POME Month']
        
        # Find the corresponding Revenue row for this POME Month first
        revenue_row_idx = None
        for rev_idx, rev_row in filled_table.iterrows():
            if (rev_row['POME Month'] == pome_month and 
                rev_row['Metric'] == 'Revenue'):
                revenue_row_idx = rev_idx
                break
        
        if revenue_row_idx is None:
            continue
            
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                filled_table.at[idx, month_col] = 0
            else:
                # Sum from the current month to the end of the analysis period
                current_month_index = month_columns.index(month_col)
                
                # Get the revenue values from the current month to the end
                revenue_values = []
                for future_month in month_columns[current_month_index:]:
                    revenue_values.append(filled_table.at[revenue_row_idx, future_month])
                
                # Calculate the cumulative sum
                cumulative_sum = sum(revenue_values)
                
                filled_table.at[idx, month_col] = int(round(cumulative_sum, 0))

# Handle Retention Rate calculation separately (Excel-like formula)
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'Retention Rate':
        pome_month = row['POME Month']
        cohort_size = row['Cohort Size']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If month column is less than POME Month, set to 0 (equivalent to "")
                filled_table.at[idx, month_col] = 0
            elif month_col == pome_month:
                # If month column equals POME Month, calculate: (SUMIFS(orders) / cohort_size) - 1
                # SUMIFS: orders where month=month_col AND pome_month=pome_month AND isFiltered=1
                filtered_data = RawData[
                    (RawData['month'] == month_col) & 
                    (RawData['pome_month'] == pome_month) & 
                    (RawData['IsFiltered'] == 1)
                ]
                
                orders_sum = filtered_data['orders'].sum()
                if cohort_size > 0:
                    retention_rate = (orders_sum / cohort_size) - 1
                    filled_table.at[idx, month_col] = round(retention_rate*100, 2)
                else:
                    filled_table.at[idx, month_col] = 0
            else:
                # If month column is greater than POME Month, calculate: Active_Customers / cohort_size
                # This corresponds to E9/$C9 in the Excel formula
                # Find the corresponding Active Customers row for this POME Month
                active_customers_row_idx = None
                for ac_idx, ac_row in filled_table.iterrows():
                    if (ac_row['POME Month'] == pome_month and 
                        ac_row['Metric'] == 'Active Customers'):
                        active_customers_row_idx = ac_idx
                        break
                
                if active_customers_row_idx is not None and cohort_size > 0:
                    active_customers = filled_table.at[active_customers_row_idx, month_col]
                    retention_rate = active_customers / cohort_size
                    filled_table.at[idx, month_col] = round(retention_rate*100, 4)
                else:
                    filled_table.at[idx, month_col] = 0

# Handle LTV calculation separately (Excel-like formula)
for idx, row in filled_table.iterrows():
    if row['Metric'] == 'LTV':
        pome_month = row['POME Month']
        cohort_size = row['Cohort Size']
        
        # For each month column
        for month_col in month_columns:
            if month_col < pome_month:
                # If month column is less than POME Month, set to 0
                filled_table.at[idx, month_col] = 0
            else:
                # Otherwise, calculate: Revenue / Cohort Size
                # Find the corresponding Revenue row for this POME Month
                revenue_row_idx = None
                for rev_idx, rev_row in filled_table.iterrows():
                    if (rev_row['POME Month'] == pome_month and 
                        rev_row['Metric'] == 'Revenue'):
                        revenue_row_idx = rev_idx
                        break
                
                if revenue_row_idx is not None and cohort_size > 0:
                    revenue = filled_table.at[revenue_row_idx, month_col]
                    ltv = revenue / cohort_size
                    filled_table.at[idx, month_col] = round(ltv, 2)
                else:
                    filled_table.at[idx, month_col] = 0

# Calculate Total column for each row
for idx, row in filled_table.iterrows():
    metric = row['Metric']
    total_value = sum([row[month] for month in month_columns])
    filled_table.at[idx, 'Total'] = total_value
```

```python
import pandas as pd

df_raw_data = RawData.copy()
df_raw_data['month'] = pd.to_datetime(df_raw_data['month'])
df_raw_data['pmonth_date'] = pd.to_datetime(df_raw_data['pome_month'])

# Create a dummy df_mom_retention based on your screenshot
mom_retention_data = {
    'Month': (list(filled_table['POME Month'])),
    'Cohort': (list(filled_table['Cohort Size']))
}

df_mom_retention = pd.DataFrame(mom_retention_data).drop_duplicates().reset_index(drop=True)
df_mom_retention['Month'] = pd.to_datetime(df_mom_retention['Month'])

def calculate_retention(cohort_month, retention_month, df_raw_data):
    """
    Calculates the MoM retention value based on the Excel formula:
    =IF(D$3<$B14,"",SUMIFS(RawData!$I:$I,RawData!$D:$D,$B14,RawData!$A:$A,D$3)/SUMIFS(RawData!$F:$F,RawData!$D:$D,$B14,RawData!$A:$A,D$3))
    
    This calculates Average Revenue Per User (ARPU) = sales/users

    Args:
        cohort_month (pd.Timestamp): The initial acquisition month of the cohort ($B14).
        retention_month (pd.Timestamp): The month for which retention is being calculated (D$3).
        raw_data_df (pd.DataFrame): The DataFrame containing the raw sales and users data.

    Returns:
        float or str: The retention value (sales/users) or an empty string if
                      retention_month is before cohort_month.
                      Returns 0 if denominator (users) is 0 to avoid division by zero.
    """

    # IF(D$3<$B14,"",...)
    if retention_month < cohort_month:
        return ""

    # Filter raw data for the specific cohort month and retention month
    # SUMIFS conditions: RawData!$D:$D=$B4 AND RawData!$A:$A=D$3
    filtered_data = df_raw_data[
        (df_raw_data['pmonth_date'] == cohort_month) &  # $D:$D=$B4 (pome_month)
        (df_raw_data['month'] == retention_month)       # $A:$A=D$3 (month)
    ]

    # SUMIFS(RawData!$I:$I,...) = sum of sales column
    sum_sales = filtered_data['sales'].sum()
    
    # SUMIFS(RawData!$F:$F,...) = sum of users column  
    sum_users = filtered_data['users'].sum()

    if sum_users == 0:
        return 0  # To handle #DIV/0! equivalent, return 0 
    else:
        return sum_sales / sum_users  # This is ARPU (Average Revenue Per User)

# --- 3. Apply the function to populate the MoM Retention table ---

# Get unique cohort months from your df_mom_retention
cohort_months = sorted(df_mom_retention['Month'].unique())

# Instead of using all months, use only the cohort analysis months (starting from 2024-07-01)
# This matches the cohort table structure
retention_columns_str = sorted(cohort_months)  # Use the same months as cohort analysis
retention_columns = [pd.to_datetime(month) for month in retention_columns_str]  # Convert to datetime for comparison


print(f"Cohort months: {[str(d) for d in cohort_months]}")
print(f"Retention columns: {retention_columns_str}")

retention_table = pd.DataFrame(index=cohort_months, columns=retention_columns)
# retention_table
for cohort_m in cohort_months:
    for retention_m in retention_columns_str:
        retention_table.loc[cohort_m, retention_m] = calculate_retention(
            cohort_m, retention_m, df_raw_data
        )

retention_table = retention_table.fillna('')
retention_table
```

```python
# Create Product LTV table exactly like the image format
def create_product_ltv():
    """Create Product LTV table matching the exact format from the image"""
    
    # Get top 6 products by cohort size (as shown in image)
    top_products = merchant_sku.sort_values('users', ascending=False).head(6)
    
    final_table = []
    
    for idx, (_, product_row) in enumerate(top_products.iterrows(), 1):
        sku = product_row['merchant_sku']
        cohort_size = product_row['users']
        product_title = product_row['pntb_title']
        
        # Get ProductRaw data for this SKU
        sku_data = ProductRaw[ProductRaw['merchant_sku'] == sku].copy()
        
        # Create the main product entry with all metrics
        product_entry = {
            '#': idx,
            'Product Title': product_title,
            'Cohort Size': cohort_size,
            'Metric': ''  # This will hold multiple metric types
        }
        
        # Add month columns (Month 1 through Month 12)
        for month_num in range(1, 13):
            month_idx = month_num - 1  # Convert to 0-based for data lookup
            month_data = sku_data[sku_data['months'] == month_idx]
            
            # Initialize all metrics for this month
            active_customers = month_data['users'].sum() if not month_data.empty else 0
            purchases = month_data['quantity'].sum() if not month_data.empty else 0  
            revenue = int(month_data['sales'].sum()) if not month_data.empty else 0
            
            # Calculate cumulative revenue
            cumulative_revenue = 0
            for cum_month in range(0, month_idx + 1):
                cum_data = sku_data[sku_data['months'] == cum_month]
                cumulative_revenue += cum_data['sales'].sum() if not cum_data.empty else 0
            cumulative_revenue = int(cumulative_revenue)
            
            # Calculate retention rate
            retention_rate = round((active_customers / cohort_size * 100), 2) if cohort_size > 0 else 0.00
            
            # Calculate LTV
            ltv = round(revenue / cohort_size, 2) if cohort_size > 0 else 0.00
            
            # Store all metrics for this month in a structured way
            product_entry[f'Month {month_num}'] = {
                'Active Customers': active_customers,
                'Purchases': purchases,
                'Revenue': revenue,
                'Cumulative Revenue': cumulative_revenue,
                'Retention Rate': retention_rate,
                'LTV': ltv
            }
        
        final_table.append(product_entry)
    
    return final_table

# Create the table
product_ltv_data = create_product_ltv()

# Display the results in a readable format
print("PRODUCT LTV ANALYSIS")
print("=" * 100)

for product in product_ltv_data:
    print(f"\n{product['#']}. {product['Product Title']}")
    print(f"   Cohort Size: {product['Cohort Size']}")
    print(f"   SKU: {merchant_sku.iloc[product['#']-1]['merchant_sku']}")
    
    # Display metrics table for this product
    metrics = ['Active Customers', 'Purchases', 'Revenue', 'Cumulative Revenue', 'Retention Rate', 'LTV']
    
    print("\n   Monthly Metrics:")
    header = "   Metric".ljust(20) + "".join([f"Month {i}".rjust(10) for i in range(1, 13)])
    print(header)
    print("   " + "-" * 140)
    
    for metric in metrics:
        row = f"   {metric}".ljust(20)
        for month_num in range(1, 13):
            value = product[f'Month {month_num}'][metric]
            if metric == 'Retention Rate':
                row += f"{value}%".rjust(10)
            elif metric == 'LTV':
                row += f"${value}".rjust(10)
            else:
                row += str(value).rjust(10)
        print(row)
    print("\n" + "=" * 100)

print(f"\nTotal products analyzed: {len(product_ltv_data)}")
```
